```python
!pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

```python
df1 = pd.read_csv('/hist_data.csv')
df1.head()
```

```python
x1=df1['aデータ']
x2=df1['bデータ']

labels =['aデータ','bデータ']
plt.hist(x1,rwidth=1.0,bins=25,alpha=0.5,label ='aデータ')
plt.hist(x2,rwidth=1.0,bins=25,alpha=0.5,label ='bデータ')
plt.title("aデータとbデータの温度に関する度数分布表")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
```

```python
df2 = pd.read_csv('/scatter_data.csv')
df2.head()
```

```python
import seaborn as sns
```

```python
x = df2['length (cm)']
y = df2['width (cm)']
z = df2['class']
sns.scatterplot(x=x,y=y,hue=z,palette=["red", "blue", "green"])

plt.title("クラス0-2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
```

```python
df3 = pd.read_csv('/content/line_data.csv')
df3.head()
```

```python
x = df3['期間']
y = df3['件数']
plt.figure(figsize=(8, 4))
plt.plot(x,y,marker='o')  
plt.grid(True)

plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
```

```python
x = df3['期間']
y = df3['費用']
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.grid(True)

plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
```
